```python
# Start writing code here...def celcius (a_string):
def celcius (a_string):
    out = a_string.replace("Â°", "")
    cel = round((int(out) - 32) * 5.0/9.0)
    return cel


def scrape ():
    import requests
    import pandas as pd
    from bs4 import BeautifulSoup

    page = requests.get("https://weather.com/weather/tenday/l/San+Francisco+CA?canonicalCityId=dfdaba8cbe3a4d12a8796e1f7b1ccc7174b4b0a2d5ddb1c8566ae9f154fa638c")
    soup = BeautifulSoup(page.content, 'html.parser')
    rows_list = []

    header = soup.findAll("details", {"summarycontentclassname": "DaypartDetails--DetailSummaryContent--1c28m"})
    for each in header:
        weatherResults = {}
        date = []
        date = each.find("h2", {"class": "DetailsSummary--daypartName--1Mebr"}).get_text(strip=True).split()
        weatherResults["Day of Week"] = date[0]
        if len(date) == 2:
            weatherResults["Day Date"] =  date[1]
        else:
            weatherResults["Day Date"] = ""
            
        weatherResults["Temp High"] = celcius(each.find("span", {"class": "DetailsSummary--highTempValue--3x6cL"}).get_text(strip=True)) # DetailsSummary--highTempValue--3x6cL
        weatherResults["Temp Low"] = celcius(each.find("span", {"class": "DetailsSummary--lowTempValue--1DlJK"}).get_text(strip=True))  # DetailsSummary--lowTempValue--1DlJK
        weatherResults["Summary"] = each.find("p", {"class": "DailyContent--narrative--3AcXd"}).get_text(strip=True)                    # DailyContent--narrative--3AcXd
        rows_list.append(weatherResults)                               
                                  
    df = pd.DataFrame(rows_list)    
    return df
scrape ()
```
